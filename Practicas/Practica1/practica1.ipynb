```python
# Tamaño de la gráfica
plt.figure(figsize=(16,6))

# Cantidad de apariciones de cada año de construcción
plot_data = data1['Años'].value_counts()

# Valores únicos (años) y sus frecuencias
x = plot_data.index
y = plot_data.values

# Gráfico de barras
plt.bar(x,y)

# Títulos de los ejes y del gráfico
plt.xlabel("Años")
plt.ylabel("Cantidad de inmuebles")
plt.title('Distribución de Años de construcción')

# Mostrar la gráfica
plt.show()
```

```python
plt.figure(figsize=(16,6))

# Variable a graficar
x = data1['Precio']

# Histograma
plt.hist(x, bins=50)

# Títulos de los ejes y de la gráfica
plt.xlabel(" Precio")
plt.ylabel("Frecuencia")
plt.title('Precio de las casas')

# Mostrar gráfica
plt.show()
```

```python
plt.figure(figsize=(10,10))

x = data1['Baños']
y = data1['Precio']

plt.plot(x, y, '.', color='blue')
plt.title('No. Baños vs Precio')
plt.xlabel('No. Baños')
plt.ylabel('Precio')
plt.show()
```

```python
# Obtener el número de valores de cada categoría
bedrooms_count = data1['Cuartos'].value_counts()#.head(5)

# Función para ilustrar en forma de pie
plt.pie(bedrooms_count, labels=bedrooms_count.index, autopct='%1.1f%%')

# Título
plt.title('Distribución por Número de Habitaciones')

# Mostrar
plt.show()
```
